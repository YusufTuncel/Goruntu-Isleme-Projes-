```python
import cv2
import matplotlib.pyplot as plt

def draw_locations(img, locations, color):
    # Check if locations is not empty
    if len(locations) > 0:
        for loc in locations:
            cv2.rectangle(img, (loc[0], loc[1]), (loc[0] + loc[2], loc[1] + loc[3]), color, 2)

def main():
    # Görüntü dosyasını yükleyin
    filename = input("Lütfen bir görüntü dosyasının yolunu girin: ")
    img = cv2.imread(filename)

    if img is None:
        print("Görüntü dosyası açılamadı.")
        return

    # HOG detektörünü başlat
    reference = (0, 255, 0)
    hog = cv2.HOGDescriptor()
    hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

    # Görüntüyü gri tonlamaya çevir
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # İnsanları algıla
    locations, _ = hog.detectMultiScale(gray)

    # Algılanan insanların etrafına dikdörtgenler çiz
    draw_locations(img, locations, reference)

    # Görüntüyü matplotlib ile göster
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Eksenleri kapat
    plt.show(/content/g.k.jpeg)

if __name__ == "__main__":
    main()
```
